```python
def compute_mean (input):
    sum = 0
    for val in input:
        sum = sum + val
    return sum/ len(input)

print (compute_mean([2, 4, 6, 8]))
```

compute_mean() calculates the mean of an input List by summing every value in the list using a for loop, then dividing it with the number of values summed.
$$
\overline{x} = 
\frac{x_1 + x_2 + ... + x_n}{n} = 
\frac{1}{n} \left( \sum_{i=1}^{n} x_i \right)
$$
